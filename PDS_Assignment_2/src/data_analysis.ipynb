```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data_clean/clean_raw_studentsperformance.csv")
df
```

```python
# Visualizing the distribution of scores by gender
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Scores by Gender')
plt.show()

sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Scores by Gender')
plt.show()

sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Writing Scores by Gender')
plt.show()


```

```python
# Visualizing the distribution of scores by race/ethnicity
sns.boxplot(x='race/ethnicity', y='math score', data=df)
plt.title('Math Scores by Race/Ethnicity')
plt.show()

sns.boxplot(x='race/ethnicity', y='reading score', data=df)
plt.title('Reading Scores by Race/Ethnicity')
plt.show()

sns.boxplot(x='race/ethnicity', y='writing score', data=df)
plt.title('Writing Scores by Race/Ethnicity')
plt.show()

```

```python
# Visualizing the distribution of scores by parental level of education
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title('Math Scores by Parental Level of Education')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='reading score', data=df)
plt.xticks(rotation=45)
plt.title('Reading Scores by Parental Level of Education')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='writing score', data=df)
plt.xticks(rotation=45)
plt.title('Writing Scores by Parental Level of Education')
plt.show()
```

```python
# Visualizing the distribution of scores by lunch
sns.boxplot(x='lunch', y='math score', data=df)
plt.title('Math Scores by Lunch')
plt.show()

sns.boxplot(x='lunch', y='reading score', data=df)
plt.title('Reading Scores by Lunch')
plt.show()

sns.boxplot(x='lunch', y='writing score', data=df)
plt.title('Writing Scores by Lunch')
plt.show()

```
